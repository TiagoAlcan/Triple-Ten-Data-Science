-	`departments.csv`
    -	`'department_id'`: é o número de identificação exclusivo de cada categoria de departamento do supermercado
    -	`'department'`: é o nome do departamento

### [A3] Em que dia da semana as pessoas compram produtos alimentícios?

```python

list_alimenticios = [1,3,4,5,6,7,9,10,12,14,15,16,19,20]


merge_orders_order_products = instacart_orders.merge(order_products, on='order_id')

merge_orders_order_products_products = merge_orders_order_products.merge(products,on='product_id')

final_df = merge_orders_order_products_products.merge(departments,on='department_id')

final_df_alimenticios = final_df[final_df['department_id'].isin(list_alimenticios)]

pedidos_alimenticios_por_dia = final_df_alimenticios.groupby('order_dow')['order_id'].nunique()

print(pedidos_alimenticios_por_dia)

pedidos_alimenticios_por_dia.plot(kind='bar')
plt.title('Pedidos de Produtos Alimentícios por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Número de Pedidos')
plt.show()
```

### [C2] Quais são os 20 principais itens incluídos com mais frequência em pedidos repetidos? Exiba os IDs e nomes.

### [C3] Para cada produto, qual parcela de todos os pedidos dele são repetidos?

```python

# Total de vezes que cada produto foi pedido
total_por_produto = order_products.groupby('product_id').size()
print(f"Exemplo - Produto 24852 foi pedido {total_por_produto[24852]} vezes")

```

```python

# Contar apenas pedidos repetidos por produto
repetidos_por_produto = order_products[order_products['reordered'] == 1].groupby('product_id').size()
print(f"Exemplo - Produto 24852 foi repetido {repetidos_por_produto[24852]} vezes")
```

```python



# Calcular proporção
proporcao_repetidos = repetidos_por_produto / total_por_produto
proporcao_repetidos = proporcao_repetidos.fillna(0)  # Produtos nunca repetidos = 0

# Criar tabela final com nomes
tabela_proporcao = proporcao_repetidos.reset_index()
tabela_proporcao.columns = ['product_id', 'proporcao_repetidos']

# Completar o merge
tabela_final = tabela_proporcao.merge(products[['product_id', 'product_name']], on='product_id')

# Ordenar por proporção decrescente para ver os produtos mais repetidos
tabela_final = tabela_final.sort_values('proporcao_repetidos', ascending=False)

# Mostrar os primeiros resultados
print("Top 20 produtos com maior proporção de repetição:")
print(tabela_final.head(20))

# Criar visualização dos top 20 produtos com maior proporção de repetição
top_20_proporcao = tabela_final.head(20)




```

```python







plt.figure(figsize=(12, 10))
plt.barh(range(len(top_20_proporcao)), top_20_proporcao['proporcao_repetidos'])
plt.yticks(range(len(top_20_proporcao)), top_20_proporcao['product_name'])
plt.xlabel('Proporção de Pedidos Repetidos')
plt.title('Top 20 Produtos com Maior Proporção de Repetição')
plt.show()




```

Escreva suas conclusões aqui

Podemos ver que esses produtos que estao sendo exibidos, sempre são repetidos, com uma taxa de repetição de 1.0

### [C4] Para cada cliente, qual proporção de todos os seus pedidos são repetidos?

```python
# instacart_orders: order_id, user_id, order_number
# order_products: order_id, product_id, reordered

ordens_produtos = instacart_orders.merge(order_products, on='order_id')


pedidos_total = ordens_produtos.groupby('user_id')['product_id'].count()

print(pedidos_total)

pedidos_repetidos = ordens_produtos[ordens_produtos['reordered'] == 1].groupby('user_id')['product_id'].count()

proporcao_pedidos = pedidos_repetidos /pedidos_total

proporcao_pedidos = proporcao_pedidos.fillna(0)

print(proporcao_pedidos)
```

```python

proporcao_pedidos.plot(kind='hist', bins= 20, title="Distribuição da proporção de pedidos repetidos")
plt.xlabel('Porporção')
plt.ylabel('Número de Clientes')
plt.show()
```

Escreva suas conclusões aqui

Podemos perceber que quando o número e 0, diz que um cliente, que fez mais de um pedido, não repetiu qualquer item que seja em seus pedidos. Além de mostrar que, normalmente, mais de 50% dos clientes repetem produtos já comprados em seus novos pedidos, uma vez que a parte do gráfico de 50% em diante é a mais populosa.
